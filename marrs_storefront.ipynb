## Crawl MarineUAS Drive and Display Raster Metadata and Footprint

Import all necessary python modules

```python
import os
from pathlib import Path
from pyproj import transform
from pyproj import Proj
import numpy as np
from shapely.geometry import shape, Point, LineString, Polygon
```

Crawl all folders for `.tif` files with `transparent_mosaic` in the filename and output those filepaths

```python
filenames = []

for fp in Path('/Volumes/Research/MarineUAS/Projects/Student Projects/').glob('**/*transparent_mosaic*.tif'):
    if 'tiles' in str(fp):
        continue
    filenames.append(fp)
    print(fp)
```

Take these filenames and open them with `rasterio` to get the footprint and necessary metadata

```python
import rasterio

gps_point_sets = []

metadata_list = []

for fp in filenames:
    try:
        with rasterio.open(fp) as dataset:
            p1 = Proj(dataset.crs.wkt) # our current crs
            p2 = Proj("+proj=longlat +ellps=WGS84 +datum=WGS84 +no_defs ") # the crs we want our raster to be in
            raster_utm_points = dataset.transform * (0, 0), dataset.transform * (dataset.width, 0), dataset.transform * (dataset.width, dataset.height), dataset.transform * (0, dataset.height)
            lon, lat = transform(p1, p2, np.array(raster_utm_points)[:,0],np.array(raster_utm_points)[:,1])
            raster_gps_points = list(zip(lat, lon)) # zip up the points so they're in the form [(lat, long), ...]
            gps_point_sets.append(raster_gps_points)

            basename = os.path.basename(fp)
            date = basename[:6]

            flight_num = basename.split("_")[0][-2:]
            sensor_type = basename.split("_")[0][-4:-2]
            location = basename.split("_")[0][6:-4]
            if not flight_num.isdigit():
                flight_num = "01"
                sensor_type = basename.split("_")[0][-2:]
                location = basename.split("_")[0][6:-2]

            metadata = {
                "crs" : dataset.crs.wkt,
                "location" : location,
                "date" : date,
                "flight_num" : flight_num,
                "sensor_type" : sensor_type,
                "filename" : fp,
                "shape" : dataset.shape,
                "size" : dataset.shape[0] * dataset.shape[1],
                "bands" : dataset.count,

            }

            metadata_list.append(metadata)
    except:
        continue

    
```

View this metadata per item

```python
for md in metadata_list:
    print(os.path.basename(md['filename']))
    print('\t', md['date'], ' - ', md['location'], ' - ', md['sensor_type'])
```

Display raster footprints

```python
# sort metadata and gps_points by size so that largest are under the smallest rasters

import folium # let's make an interactive map using leaflet

# creating these points just to set the map center
lat, long = gps_point_sets[0][0] 

# create the folium map object
m = folium.Map(location=[lat, long], zoom_start=2) # set the map centered around the first point

for idx, footprint in enumerate(gps_point_sets):
    # this actually adds the polygon to the map
    folium.Polygon(footprint, 
                   popup="".join([metadata_list[idx]['location'], " - ", metadata_list[idx]['sensor_type'], 
                                  " - ",  metadata_list[idx]['date']]),
                   color='#3186cc',
                   fill=True,
                   fill_color='#3186cc'
                  ).add_to(m)
    #print(os.path.basename(metadata_list[idx]['filename']))

m
```
